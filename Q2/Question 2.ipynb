### An overview of how the problem should be solved.
- Main task is to assign the right vaccine types and amount to the vaccination centres.
- import constraint
- variable -> vaccine center
- domain -> range
- constraint -> number of people
- feed in the vaccine centres and the number to the systems, feed in the maximum number of vaccine centers per state, feed in the maximum number of vaccines per day per state
- goal -> return combination with exact sum of maximum number of vaccines per day per state ( basically all the possible combination from that combination)(interate through to find the cheapest cost)
- seperate into two which is :-last day = population % daily doses & everyday except last day = population - last day.

### Different ways to formulate and present the problem.
- To solve the problem, we can formulate and present the problem by choosing which group to vaccinates first.Let old generation first since the smallest number and then yound generation and middle age.
    
### What are the important functions in CSP?
- A finite variable set, a domain set and a finite constraint set

### What are the parameters to be optimized?
- Parameter to be optimized are constraints

### What are the criteria to rank the solutions?
- Criteria to rank the solutions are:-
- 1.Find the fastest possible solution which is the minimum days needed to finish vaccine people in that state.
- 2.Find the cheapest possible solution which is the cheapest cost for the vaccine centre rent.

### How the assess the fitness of the solutions?
- I assess the fitness by speed and cost for the solution.I calculate the minimum days needed to finish vaccine people in that state and then calculate the cheapest cost for the vaccine centre rent.

## Cheapest Cost For 5000 Vaccines and calculation to determine last day

```python
state1_pb = constraint.Problem()

state1_pb.addVariable("CR-1", range(21)) # 20 vaccination_centre1
state1_pb.addVariable("CR-2", range(16)) # 15 vaccination_centre2
state1_pb.addVariable("CR-3", range(11)) # 10 vaccination_center3
state1_pb.addVariable("CR-4", range(22)) # 21 vaccination_center4
state1_pb.addVariable("CR-5", range(6)) # 5 vaccination_center5

state1_pb.addConstraint(
    constraint.ExactSumConstraint(5000,[200,500,1000,2500,4000]),
    ["CR-1", "CR-2", "CR-3", "CR-4", "CR-5"])

state1_cheapest_cost = math.inf 
state1_solfound = {}
state1_sol = state1_pb.getSolutions()

for s in state1_sol:
    cheapest_cost = s["CR-1"]*100 + s["CR-2"]*250 + s["CR-3"]*500 +s["CR-4"]*800 + s["CR-5"]*1200
    if cheapest_cost < state1_cheapest_cost:
        state1_cheapest_cost =  cheapest_cost
        state1_solfound= s
    
print("""  
The cheapest cost for 5000 vaccines is : {}  
We'll use:  
{} CR-1,
{} CR-2,
{} CR-3,
{} CR-4,
{} CR-5
""".format(state1_cheapest_cost, state1_solfound["CR-1"], state1_solfound["CR-2"], state1_solfound["CR-3"],
           state1_solfound["CR-4"], state1_solfound["CR-5"]))

#vA is Age > 60, vB is Age between 35 to 60 and vC Age < 35
#maxV is maximum people in that state can vaccine per day
vA = 15000 
vB = 434890
vC = 115900
maxV = 5000
(vA + vB + vC) % maxV
```

```python
state1_pb2 = constraint.Problem()

state1_pb2.addVariable("CR-1", range(21)) # 20 vaccination_centre1
state1_pb2.addVariable("CR-2", range(16)) # 15 vaccination_centre2
state1_pb2.addVariable("CR-3", range(11)) # 10 vaccination_center3
state1_pb2.addVariable("CR-4", range(22)) # 21 vaccination_center4
state1_pb2.addVariable("CR-5", range(6)) # 5 vaccination_center5

def custom_constraint(CR1, CR2, CR3, CR4, CR5):
    if (CR1*200 + CR2*500 + CR3*1000 + CR4*2500 + CR5*4000) >= 790:
        return True

state1_pb2.addConstraint(custom_constraint,["CR-1","CR-2","CR-3","CR-4","CR-5"])
    
state1_solfound2 = {}
state1_sol2 = state1_pb2.getSolutions()
state1_cheapest_cost2 = math.inf 

for s in state1_sol2:
    cheapest_cost = s["CR-1"]*100 + s["CR-2"]*250 + s["CR-3"]*500 +s["CR-4"]*800 + s["CR-5"]*1200
    if cheapest_cost < state1_cheapest_cost2:
        state1_cheapest_cost2 = cheapest_cost
        state1_solfound2 = s
        
print("""
The cheapest cost for 790 vaccines is : {}  
We'll use:  
{} CR-1,
{} CR-2,
{} CR-3,
{} CR-4,
{} CR-5
""".format(state1_cheapest_cost2, state1_solfound2["CR-1"], state1_solfound2["CR-2"], state1_solfound2["CR-3"],
        state1_solfound2["CR-4"], state1_solfound2["CR-5"]))
```
